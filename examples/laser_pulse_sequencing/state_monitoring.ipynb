```python
import numpy as np
import matplotlib.pyplot as plt
from qdlutils.experiments.laser_pulse_sequencing.state_monitoring import SimpleStateMonitoringExperiment
from qdlutils.hardware.wavemeters.coherentwavemaster import CoherentWaveMaster
```

```python
# Get the wavemeter controller
wavemeter_controller = CoherentWaveMaster(
    port = 'COM1',
    timeout = 2,
    units = 'F',
    mode = 'C',
    autocal = False
)
```

```python
exp = SimpleStateMonitoringExperiment(
    repump_id = 'green_laser',
    repump_do_config = {'device': 'Dev1', 'port': 'port0', 'line': 'line0'},
    probe_id = 'red_laser',
    probe_do_config = {'device': 'Dev1', 'port': 'port0', 'line': 'line1'},
    probe_ao_config = {'device': 'Dev1', 'channel': 'ao3', 'min': -3, 'max': 5},
    counter_id = 'spcm_rate',
    counter_ci_config = {'device': 'Dev1', 'channel': 'ctr2', 'terminal': 'PFI0'},
    wavemeter_controller = wavemeter_controller,
    clock_device = 'Dev1',
    clock_channel = 'ctr0'
)
```

```python
exp.set_probe_switch(True)
```

```python
exp.set_repump_switch(True)
```

```python
exp.set_probe_voltage(0)
```

```python
exp.single_probe_scan(
    voltage_min = -3,
    voltage_max = 5,
    n_pixels = 100,
    scan_time = 5,
    repump_time = 1,
    optimize = 'max'
)
```

```python
exp.stabilize(
    tol = 0.2,
    hold_window = 5,
    max_attempts = 50,
    query_period = 0.2,
    freq_volt_grad = -24,
    penalty = 0.95
)
```

```python
scan_kwargs = {
    'voltage_min': 0,
    'voltage_max': 2,
    'n_pixels': 100,
    'scan_time': 2,
    'repump_time': 1,
    'optimize': 'max'
}

exp.run(
    n_batches = 10,
    n_repetitions = 5,
    repump_time = 0.05,
    probe_delay = 0.01,
    probe_time = 0.1,
    end_delay = 0.05,
    clock_rate = 1000,
    scan_kwargs = scan_kwargs,
    stabilization_kwargs = None,
)
```
